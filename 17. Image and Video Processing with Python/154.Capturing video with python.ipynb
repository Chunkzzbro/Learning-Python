```python
import cv2,time
#DISPLAYS ONLY THE FIRST FRAME:

#Switching webcame to take video : 

video = cv2.VideoCapture(0)                         #Switches on the web camera. 0 for builtin camera and 1 for external camera
check,frame = video.read()                          #Captures Each frame and also checks whether video is read or not 
print(check)                                        #Prints whether true or false with respect to if video is read or not
print(frame)                                        #Print Frame in the form of a numpy array

#Converting frame captured to grayscale (Optional):

gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)       #Converts the BGR frame into a grayscale
#time.sleep(3)                                      #Makes the control of python program sleep for 3 seconds

#Displaying the captured frame in an external window:

cv2.imshow("Capturing",gray)                        #Shows the converted gray scale image in another window
cv2.waitKey(0)                                      #Transfers control to next statement when a key is pressed form the window
video.release()                                     #The webcame turns off and video is released
cv2.destroyAllWindows()                             #Exteral window gets destroyed






 
```

```python
#Displaying an entire video :

video = cv2.VideoCapture(0)                         #Switches on the web camera. 0 for builtin camera and 1 for external camera

frame_number = 0

while True:                                                   #Iterating per milli second to display a video
    frame_number = frame_number + 1
    check,frame = video.read()                                #Captures Each frame and also checks whether video is read or not 
    print(check)                                              #Prints whether true or false with respect to if video is read or not
    print(frame)                                              #Print Frame in the form of a numpy array

    #Converting frame captured to grayscale (Optional):

    #gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)             #Converts the BGR frame into a grayscale
    #time.sleep(3)                                            #Makes the control of python program sleep for 3 seconds

    #Displaying the captured frame in an external window:

    cv2.imshow("Capturing",frame)                              #Shows the converted gray scale image in another window
    key = cv2.waitKey(1)                                      #Transfers control to next loop unless q is pressed where it breaks from loop
    if key==ord('q'):
        break

video.release()                                               #The webcame turns off and video is released
cv2.destroyAllWindows()                                       #Exteral window gets destroyed
print(frame_number)
    


```
