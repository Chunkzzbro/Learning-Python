```python
import folium
import pandas

data = pandas.read_csv("Volcanoes.txt")
data
```

```python
lat = list(data ['LAT'])
lon = list(data ['LON'])
elev = list(data['ELEV'])
```

```python
#Adding markers from file volcanoes.txt 

map = folium.Map(location = [39.86546881629921, -105.00650100494674],zoom_start = 5,tiles = "Stamen Terrain")
fg = folium.FeatureGroup(name = "Watashi no map")     

for lt,ln,el in zip(lat,lon,elev):
    fg.add_child(folium.Marker(location = [lt,ln],popup = el, icon = folium.Icon(colour = "green")))

map.add_child(fg)
map.save('Map2.html')

map
```
