```python
import folium
import pandas

data = pandas.read_csv("Volcanoes.txt")
data
```

```python
lat = list(data ['LAT'])
lon = list(data ['LON'])
elev = list(data['ELEV'])
```

```python
#Defining a function for presenting various colours for different elevation values


def color_producer(elevation):
    if elevation<1000:
        return 'green'
    elif 1000 <= elevation <3000:
        return 'orange'
    else:
        return 'red'
```

```python
#Adding markers from file volcanoes.txt along with their elevation

map = folium.Map(location = [39.86546881629921, -105.00650100494674],zoom_start = 5,tiles = "Stamen Terrain")
fg = folium.FeatureGroup(name = "Watashi no map")     

for lt,ln,el in zip(lat,lon,elev):
    fg.add_child(folium.Marker(location = [lt,ln],popup = str(el) + "m", icon = folium.Icon(color = color_producer(el))))

map.add_child(fg)
map.save('Map3.html')

map
```
