```python
import folium
import pandas
```

```python
data = pandas.read_csv("Volcanoes.txt")
data
```

```python
lat = list(data['LAT'])
lon = list(data['LON'])
elev = list(data['ELEV'])
```

```python
def color_producer(elevation):
    if elevation <1000:
        return 'green'
    elif elevation >= 1000 and elevation < 3000:
        return 'orange'
    else:
        return 'red'




```

```python
map = folium.Map(location = [39.86546881629921, -105.00650100494674],zoom_start = 5,tiles = "Stamen Terrain")
fg = folium.FeatureGroup(name = "Watashi no map")

for lt,ln,el in zip (lat,lon,elev):
    fg.add_child(folium.CircleMarker(location = [lt,ln] ,radius = 6, popup = str(el)+"m" , fill_color = color_producer(el),color = 'grey',fill_opacity = 0.7))
    
map.add_child(fg)
map
```
