```python
import folium
import pandas

data = pandas.read_csv("Volcanoes.txt")
data


```

```python
lat = list(data ['LAT'])
lon = list(data ['LON'])
print (lat)
print (lon)
```

```python
#Adding markers from file volcanoes.txt 

map = folium.Map(location = [39.86, -105.00],zoom_start = 5,tiles = "Stamen Terrain")
fg = folium.FeatureGroup(name = "Watashi no map")     

for lt,ln in zip(lat,lon):
    fg.add_child(folium.Marker(location = [lt,ln],popup = "Hi i am a marker", icon = folium.Icon(colour = "green")))

map.add_child(fg)

map
```
