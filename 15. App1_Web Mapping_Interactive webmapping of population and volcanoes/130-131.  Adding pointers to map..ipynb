```python
import folium
map = folium.Map(location = [8.52,76.53],zoom_start = 6,tiles = "Stamen Terrain")
map


```

```python
#Adding a point marker

fg = folium.FeatureGroup(name = "Watashi no map")                                       #To keep your codes organized

fg.add_child(folium.Marker(location = [8.6,76.6],popup = "Hi i am a marker", icon = folium.Icon(colour = "green")))
map.add_child(fg)
map
              

```

```python
#Adding multiple markers :
# 1. Copy pasting code with different locations
fg.add_child(folium.Marker(location = [7.6,75.6],popup = "Hi i am a marker", icon = folium.Icon(colour = "green")))
map


```

```python
# 2. Using a for loop:

for coordinates in [[7.6,75.6],[8.6,76.6]]:
    fg.add_child(folium.Marker(location = coordinates,popup = "Hi i am a marker", icon = folium.Icon(colour = "green")))

map
```
