```python
from motion_detector import df
from bokeh.plotting import figure,output_file,show
from bokeh.models import HoverTool, ColumnDataSource                   #Column data source is the standard mode of giving data to  bokeh plot.

df["Start_string"] = df["Start"].dt.strftime("%y-%m-%d %H:%M:%S")      #Converting the date times to year-month-day-hour-minute-second format
df["End_string"] = df["End"].dt.strftime("%y-%m-%d %H:%M:%S")

cds = ColumnDataSource(df)

p = figure(height = 500, width = 500,x_axis_type = "datetime")
p.title.text = "Motion Graph"
q = p.quad(left = "Start", right = "End",bottom = 0, top = 1, color = "green", source = cds)

#Adding a HoverTool to the graph so that when you hover your cursor over the graph it gives back the time of start and end.
hover = HoverTool(tooltips = [("Start","@Start_string"),("End","@End_string ")])
p.add_tools(hover)                                                        #Adding it to the bokeh plot.


output_file("Graph.html")

show (p)
```
