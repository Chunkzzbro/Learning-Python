```python
import requests
from bs4 import BeautifulSoup 


r = requests.get ("http://pythonizing.github.io/data/real-estate/rock-springs-wy/LCWYROCKSPRINGS/")
c = r.content


#Beautifying the source code that is extracted into c variable
soup = BeautifulSoup(c,"html.parser")


#Extracting div elements:

all = soup.find_all("div",{"class":"propertyRow"})


#Extracting seperate elements:

all[0].find("h4",{"class":"propPrice"}).text.replace("\n","").replace(" ","")

#Replacing the extra spaces and unwanted items with nothing
```

```python


for item in all:
    
    
    #Extracting Cost of properties with a for loop
    print(item.find("h4",{"class","propPrice"}).text.replace("\n","").replace(" ",""))
   

    #Extracting Address
    
    print(item.find_all("span",{"class","propAddressCollapse"})[0].text)
    print(item.find_all("span",{"class","propAddressCollapse"})[1].text)
    print("")
    
    #Extracting Bed number
    
    try :
        print(item.find("span",{"class","infoBed"}).find("b").text)
        
    except :
        print(None)                                                                         #An error might occur for the datas with no bed type entered since the datatype will be none
    
    
    #Extracting Area Size
    
    try :
        print(item.find("span",{"class","infoSqFt"}).find("b").text)
        
    except :
        print(None)    
    
    #Extracting full bath
    
    try :
        print(item.find("span",{"class","infoValueFullBath"}).find("b").text)
        
    except :
        print(None)    
    
    #Extracting Half bath
    
    try :
        print(item.find("span",{"class","infoValueHalfBath"}).find("b").text)
        
    except :
        print(None)   
    
    print("")
```
